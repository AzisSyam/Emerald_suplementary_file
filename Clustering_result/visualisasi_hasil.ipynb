```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def visualize_clustering_result(data, kmeans, title):
    plt.figure(figsize=(6, 6))
    scatter = plt.scatter(data['ZS TB/U'], data['ZS BB/U'], c=kmeans, cmap='viridis')
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100, label='Centroids')
    plt.xlabel('z-score height for age')
    plt.ylabel('z-score weight for age')
    plt.title(title)
    # Custom legend untuk setiap klaster
    plt.legend(handles=scatter.legend_elements()[0], labels=[f'Cluster {i+1}' for i in range(max(kmeans.labels_)+1)] + ['Centroids'])
    plt.grid(True)
    plt.show()
```

```python
clustering_usia_0 = pd.read_excel('clustering_usia_0.xlsx')
clustering_usia_1 = pd.read_excel('clustering_usia_1.xlsx')
clustering_usia_2 = pd.read_excel('clustering_usia_2.xlsx')
clustering_usia_3 = pd.read_excel('clustering_usia_3.xlsx')
clustering_usia_4 = pd.read_excel('clustering_usia_4.xlsx')
```

```python
visualize_clustering_result(clustering_usia_0, clustering_usia_0['cluster'], 'Results of 0-Year Age Clustering')
```
